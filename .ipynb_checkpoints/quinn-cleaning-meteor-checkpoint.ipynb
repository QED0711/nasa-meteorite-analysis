```python
import numpy as np 
import pandas as pd
import pdb

from classifications import gen_classifiers
```

```python
df = pd.read_csv("Meteorite_Landings.csv")
df.columns = ['name', 'id', 'nametype', 'recclass', 'mass', 'fall', 'year',
       'reclat', 'reclong', 'GeoLocation']
```

### Drop mass NaNs & 0.0

```python
df = df[(df.mass > 0.0) & (df.mass.isna() == False)]
df.shape
```

## Classifications

```python
# A function to quickly do manual search for classification type
def find_classification(classification=''):
    for c in df.recclass:
        if classification in c:
            print(c)

find_classification("Mesos")
```

```python
gen_classifiers()
```

```python
classifications = []
match_on = []

classifiers = gen_classifiers()
for i, rc in enumerate(df.recclass):
    classifications.append("uncategorized")
    match_on.append(None)
    for subclass in classifiers:
        if subclass in rc:
            classifications[i] = classifiers[subclass]
            match_on[i] = subclass
        
len(classifications)

df["major_classification"] = classifications
df['classification_match_on'] = match_on
df[['recclass', 'classification_match_on', 'major_classification']]
```
